```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

# Fill remaining missing values in x with the median
for col in x.columns:
    if x[col].dtype in [np.float64, np.int64]:
        x[col] = x[col].fillna(x[col].median())


xtrain, xtest, ytrain, ytest = train_test_split(
    x,y, random_state=42, test_size= 0.2
)

models = [RandomForestRegressor(), DecisionTreeRegressor(), LinearRegression()]

errors = {}

for model in models:
    model.fit(xtrain,ytrain)
    name = type(model).__name__
    score = model.score(xtest,ytest)*100
    errors[name] = score

display(errors)
```

```python
from sklearn.metrics import mean_squared_error as mse

# Get the Linear Regression model from the models list
lr_model = models[2] # Assuming LinearRegression is the third model in the list

predicted = lr_model.predict(xtest)
error = mse(ytest, predicted) ** .5
print(error)
```
